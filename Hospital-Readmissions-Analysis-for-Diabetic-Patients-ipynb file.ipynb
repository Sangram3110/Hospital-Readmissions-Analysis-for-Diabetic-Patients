```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r"C:\Users\Vikram\Desktop\Geekster\Projects/diabetic_data.csv")
```

```python
data.head()
```

```python
#Check rows and columns
data.shape
```

```python
#Check Null values
data.isnull().sum()
```

```python
data['max_glu_serum'].fillna('None', inplace=True)
data['A1Cresult'].fillna('None', inplace=True)
```

```python
#check columns after handelling null/missing values
data.isnull().sum()
```

```python
#see all the coluns in list
data.columns.tolist()
```

```python
# Check unique values in some important columns
important_cols = ['race', 'gender', 'age','readmitted']
for col in important_cols:
    print(f"\nUnique values in '{col}':\n", data[col].unique())
```

```python
import numpy as np
## Replace '?' with NaN
data.replace('?', np.nan, inplace=True)
```

```python
data.isnull().sum()
```

```python
# if missing % is greater than 40 then drop the rows
missing_percentage=data.isnull().mean()*100
col_to_drop=missing_percentage[missing_percentage>40].index.tolist()
print(col_to_drop)
data.drop(columns=col_to_drop,axis=1,inplace=True)
```

```python
## Drop rows with critical missing data
data.dropna(subset=['race','diag_1','diag_2','diag_3','age','payer_code'], inplace=True)
```

```python
data.isnull().sum()
```

```python
#check rows and columns after cleaning
data.shape
```

```python
#check duplicate values
data.duplicated().sum()
```

```python
data=data[data['gender'] != 'Unknown/Invalid']
```

```python
# Assuming the target column is named 'readmitted'
sns.countplot(x='readmitted', data=data)
plt.title('Distribution of Readmission')
plt.show()
```

```python
#Outlier Detection (only numeric columns)
numeric_cols = data.select_dtypes(include=[np.number]).columns
```

```python
from sklearn.ensemble import IsolationForest
# Use Isolation Forest for multivariate outlier detection
iso = IsolationForest(contamination=0.01)
outliers = iso.fit_predict(data[numeric_cols])
data= data[outliers == 1]  # Keep only inliers
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#Encode Categorical Variables using Label Encoding  (0-Readmitted within 30 days,1-Readmitted after 30 days,2- Not readmitted)

categorical_cols = data.select_dtypes(include='object').columns
le = LabelEncoder()
for col in categorical_cols:
    data[col] = le.fit_transform(data[col])
```

```python
# Readmission Count
sns.countplot(x='readmitted', data=data)
plt.title('Readmission Distribution')
plt.xlabel('Readmitted')
plt.ylabel('Count')
plt.show()
```

```python
data['readmitted'].value_counts().sort_index()
```

```python
# Gender vs Readmission
sns.countplot(x='gender', hue='readmitted', data=data)
plt.title('Gender vs Readmission')
plt.show()
```

```python
data['gender'].value_counts()
```

```python
# Age vs Readmission (Boxplot)
sns.boxplot(x='readmitted', y='age', data=data)
plt.title('Age vs Readmission')
plt.show()
```

```python
# Time in Hospital vs Readmission
sns.barplot(x='readmitted', y='time_in_hospital', data=data)
plt.title('Time in Hospital vs Readmission')
plt.show()
```

```python
# Number of Medications vs Readmission
sns.barplot(x='readmitted', y='num_medications', data=data)
plt.title('Number of Medications vs Readmission')
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(16,10))
sns.heatmap(data.corr(), cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Time in Hospital vs Readmission
sns.barplot(y='readmitted', x='insulin', data=data)
plt.title('Time in Hospital vs Readmission')
plt.show()
```
